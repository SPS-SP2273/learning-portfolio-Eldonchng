```python
from matplotlib import pyplot as plt
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color="red", linestyle="dashed",
             marker=".", markerfacecolor="yellow",
             ecolor="black",capsize=5,
             label="Y value with errors", markersize = 10 ) #capsize is the horizontal lines at the ends of the error bar.
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc = "lower right")
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
x_array = np.linspace(0,50,500)
y_array= np.exp((-1*x_array)/10) * np.sin(x_array)
plt.plot(x_array, y_array, label='Y values',
         color='blue', linestyle='solid',alpha = 1) #alpha control the opacity of the line
plt.xlabel('x', color ="darkorange", fontsize =15 )
plt.ylabel('y', color = "darkorange",fontsize = 15)
plt.title(r'y = $e^{-x/10}\sin(x)$', fontsize = 20)
plt.grid(alpha=.25)

plt.ylim([-1,1])
```
